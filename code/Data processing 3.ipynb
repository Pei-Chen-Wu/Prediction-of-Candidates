# Facenet前處理

```python
import pandas as pd
df = pd.read_csv("103-two_compare.csv")  
df
```

```python
df["編碼"] = df["編碼"].apply(str)  + '.png'
```

## 拿掉兩人中有連任的組別

```python
df_ = df.groupby(["地區"]).是否現任.describe()
df_ 
```

```python
df['unique'] = df['地區'].map(dict(zip(df_.index,df_['unique'])))
df
```

```python
df_ans = df[df['unique']==1]
df_ans
```

```python
df_ans = df_ans[['編碼','當選註記']]
df_ans
```

## 切割train dev

```python
df_ans.rename(columns={'編碼':'image','當選註記':'label'},inplace=True)
df_ans
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_ans, test_size=0.2)
```

```python
print('train筆數：',len(train))
print('test筆數：',len(test))
print('train label為0筆數：',len(train[train['label'] == 0]))
print('train label為1筆數：',len(train[train['label'] == 1]))
print('test label為0筆數：',len(test[test['label'] == 0]))
print('test label為1筆數：',len(test[test['label'] == 1]))
```

```python
train.to_csv("facenet_train.csv",index=False)
test.to_csv("facenet_test.csv",index=False)
```
